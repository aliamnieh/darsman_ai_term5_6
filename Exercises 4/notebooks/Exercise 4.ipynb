````python
# TODO: ุดูุงุณุง ุฏุงุฏู ูุง ูพุฑุช ุจู ฺฉูฺฉ IQR

numeric_columns=df.select_dtypes(include=['number']).columns
for column in numeric_columns:
    q1=np.percentile(df[column].dropna(),25)
    q3=np.percentile(df[column].dropna(),75)
    iqr=q3-q1
    lower_bound=q1-1.5*iqr
    upper_bound=q3+1.5*iqr
    outlier_amount=((df[column]<lower_bound)|(df[column]>upper_bound)).sum()
    print(f'outliers for {column} : {outlier_amount}')
````
